```python
data=pd.read_csv("company_sales_data.csv",usecols=['facewash','facecream','shampoo','toothpaste','bathingsoap','moisturizer'])
print(data)
```

```python
l=[]
for i in data:
  a=list(data[i])
  b=np.array(a)
  sum=b.sum()
  l.append(sum)
print(l)
```

```python
plt.figure(figsize=(10,10))
plt.pie(l,labels=list(data),autopct='%1.1f%%')
```
